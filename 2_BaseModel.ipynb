```python
### import libraries
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
import datetime
from datetime import timedelta
from time import time
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.cross_decomposition import PLSRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor, ExtraTreesRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error, explained_variance_score, mean_absolute_percentage_error
import warnings
warnings.filterwarnings("ignore")
```

```python
# Data extraction
data = pd.read_csv('D:/Kaggle Datasets/playground-series-s3e20/train.csv')
# top 5 rows
data[:5]
```

```python
# data split
## Train-Valid
### Train-Test

trainData, validData = train_test_split(data,test_size=0.2,random_state=42)
trainData, testData = train_test_split(trainData,test_size=0.1,random_state=42)

print(f'Train Dataset:', trainData.shape)
print(f'Valid Dataset:',validData.shape)
print(f'Test Dataset:', testData.shape)
```

```python
columnsToDrop = {'ID_LAT_LON_YEAR_WEEK','UvAerosolLayerHeight_aerosol_height','UvAerosolLayerHeight_aerosol_pressure',
                          'UvAerosolLayerHeight_aerosol_optical_depth','UvAerosolLayerHeight_sensor_zenith_angle',
                          'UvAerosolLayerHeight_sensor_azimuth_angle','UvAerosolLayerHeight_solar_azimuth_angle',
                          'UvAerosolLayerHeight_solar_zenith_angle'}
```

```python
# trainTarget = trainData['emission']
trainData.drop(columns = columnsToDrop,inplace=True, axis=1)
validData.drop(columns = columnsToDrop,inplace=True, axis=1)
testData.drop(columns = columnsToDrop,inplace=True, axis=1)
```

```python
trainData[['year','week_no']] = trainData[['year','week_no']].astype('str')
validData[['year','week_no']] = validData[['year','week_no']].astype('str')
testData[['year','week_no']] = testData[['year','week_no']].astype('str')
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
x_train = trainData.drop('emission',axis=1).copy()
y_train = trainData['emission'].copy()

x_valid = validData.drop('emission',axis=1).copy()
y_valid = validData['emission'].copy()

x_test = testData.drop('emission',axis=1).copy()
y_test = testData['emission'].copy()
```

```python
numerical = x_train.select_dtypes('number').columns.tolist()
categories = x_train.select_dtypes('object').columns.tolist()
```

```python
categories
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
numbers = Pipeline(steps=[('SimpleImpute',SimpleImputer(strategy='mean')),('scale',StandardScaler())])
cat = OneHotEncoder(handle_unknown='ignore')
```

```python
preProcessor_numbers = ColumnTransformer(transformers=[('num',numbers,numerical)
                                                ])
```

```python
preProcessor_cat = ColumnTransformer(transformers=[('Cat',cat,categories)])
```

```python
preProcessor_numbers.fit(x_train)
preProcessor_cat.fit(x_train)
cat_columns = (preProcessor_cat.get_feature_names_out()).tolist()
```

```python
def transformation(transformer1, transformer2, df):
    num_transf = pd.DataFrame(transformer1.transform(df),columns= numerical)
    cat_transf = pd.DataFrame((transformer2.transform(df)).toarray(),columns=cat_columns)

    processed = pd.concat([num_transf,cat_transf],axis=1)
    return processed
```

```python
x_train_processed = transformation(preProcessor_numbers,preProcessor_cat,x_train)
x_valid_processed = transformation(preProcessor_numbers,preProcessor_cat,x_valid)
x_test_processed = transformation(preProcessor_numbers,preProcessor_cat,x_test)
```

```python
x_train_processed.head()
```

```python
# Base Model
```

```python
regressors = [LinearRegression(), Lasso(), Ridge(), ElasticNet(), PLSRegression(), 
              KNeighborsRegressor(), GradientBoostingRegressor(), ExtraTreesRegressor(),
              RandomForestRegressor(), DecisionTreeRegressor(), LGBMRegressor(), XGBRegressor()]
```

```python
for model in regressors:
    start = time()

    model.fit(x_train_processed,y_train)
    train_time = time() - start 

    start = time()
    y_pred = model.predict(x_valid_processed)
    pred_time = time() - start 

    print(model)
    print('\tTraing time: %0.3fs' % train_time)
    print('\tPrediction time: %0.3fs' % pred_time)
    print("\tExplained variance:", explained_variance_score(y_valid, y_pred))
    print('\tRMSE: %f' % np.sqrt(mean_squared_error(y_valid, y_pred)))
    print("\tMean absolute error:", mean_absolute_error(y_valid, y_pred))
    print("\tR2 score:", r2_score(y_valid, y_pred))
    print()
```
